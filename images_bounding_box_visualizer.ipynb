```python
images_path = '/content/drive/My Drive/Tree in Forest/Supermarket shelves/images/'
ann_path = '/content/drive/My Drive/Tree in Forest/Supermarket shelves/annotations/'
```

Visualizing image 001 - replace with desired image to be visualized (options range from 001 to 045)

```python
image_name = '001.jpg'
image_path = images_path + image_name
image = cv2.imread(image_path)

# Load the JPG JSON file
json_path = ann_path + image_name + '.json'
with open(json_path, 'r') as f:
    data = json.load(f)
```

```python
for obj in data['objects']:
    # Get the exterior points of the bounding box
    pt1 = tuple(obj['points']['exterior'][0])
    pt2 = tuple(obj['points']['exterior'][1])

    # Draw the rectangle on the image
    cv2.rectangle(image, pt1, pt2, (0, 255, 0), 2)  # Green box with thickness 2
```

```python
# Convert the image from BGR (OpenCV format) to RGB (Matplotlib format)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image with bounding boxes
plt.figure(figsize=(10, 10))
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```
